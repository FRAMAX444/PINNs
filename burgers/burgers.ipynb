```python
x = np.linspace(-1, 1, 256, dtype=np.float32).squeeze()
t = np.linspace(0, 0.99, 100, dtype=np.float32).squeeze()
```

# 1 test

```python
def u_true_func(x):
    return -np.sin(np.pi * x)

u0 = u_true_func(x)

ps = PINNSolver([2, 50, 50, 50, 1], x, t, u0)

ps.plot_initial_data()
ps.plot_training_points_1D()
ps.Train()
ps.plot_solution()
ps.plot_predicted_slices_grid()
```

# 2 test

```python
def double_peak(x):
    peak1 = 1.5 * np.exp(-((x + 0.1) ** 2) / 0.02)  
    peak2 = 1.0 * np.exp(-((x - 0.3) ** 2) / 0.01) 
    return peak1 + peak2

u0_1 = double_peak(x)

ps1 = PINNSolver([2, 50, 50, 50, 1], x, t, u0_1)
ps1.plot_initial_data()
ps1.Train()
ps1.plot_solution()
ps1.plot_predicted_slices_grid()
```
